```python
import re
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords = ["https", "co", "RT", "nhttps","http","al","de"] + list(set(stopwords.words('german')))
```

```python
db_AfD_CDU = db_AfD_CDU.reset_index(drop=True)
replace_space = re.compile('[/(){}\[\]\|@,;]')
replace_symbols = re.compile('[^0-9a-z #+_]')
stopwords = stopwords

def clean_text(text):
    """
        text: a string
        
        return: modified initial string
    """
    text = text.lower() # lowercase text
    text = replace_space.sub('', text) # replace REPLACE_BY_SPACE_RE symbols by space in text. substitute the matched string in REPLACE_BY_SPACE_RE with space.
    text = replace_symbols.sub('', text) # remove symbols which are in BAD_SYMBOLS_RE from text. substitute the matched string in BAD_SYMBOLS_RE with nothing. 
    text = text.replace('x', '')
    text = ' '.join(word for word in text.split() if word not in stopwords) # remove stopwors from text
    return text
db_AfD_CDU['text'] = db_AfD_CDU['text'].apply(clean_text)
db_AfD_CDU['text'] = db_AfD_CDU['text'].str.replace('\d+', '')
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
# The maximum number of words to be used. (most frequent)
max_top_words = 50000
# Max number of words in each complaint.
max_tweet_lenght = 300

tokenizer = Tokenizer(num_words=max_top_words, filters='!"#$%&()*+,-./:;<=>?@[\]^_`{|}~', lower=True)
```

```python
tokenizer.fit_on_texts(db_AfD_CDU['text'].values)
word_index = tokenizer.word_index
print('Found %s unique tokens.' % len(word_index))
```

```python
X = tokenizer.texts_to_sequences(db_AfD_CDU['text'].values)
X = pad_sequences(X, maxlen=max_tweet_lenght)
print('Shape of data tensor:', X.shape)
```

```python
Y = pd.get_dummies(db_AfD_CDU['ideology']).values
print('Shape of label tensor:', Y.shape)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.10, random_state = 42)
print(X_train.shape,Y_train.shape)
print(X_test.shape,Y_test.shape)
```
